```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt


def perceptron_ans(train,trainlabel):
    """ PERCEPTRON performs perceptron learning algorithm for binary
     classification (online version).
     TRAIN provides the training data for the perceptron. It is a n by d
     matrix. n is the number of samples and d is the dimension of the data.
     TRAINLABEL provides the label (-1 or 1) for TRAIN. It is an n-by-1 vector.
     The function outputs weights (w(1) is the weight for the bias unit). It
     is a d+1 by 1 vector.
     OUTPUT: "w", the weights of the perceptron."""

    #get dimensions of data matrix
    n = train.shape[0]
    d = train.shape[1]

    #appending the bias coefficient (1) to each data point
    train = np.concatenate([np.ones((n,1)), train],axis=1)
    #this is the new adjusted X

    #--------Your code below-----------

    #Hint: you might find numpy's np.random.permutation function to be useful
    
    w = np.ones((d+1))

    # Controls the maximum number of iterations
    for maxiter in range(100):
#         print(maxiter, "newiteration-----------------------------------------")

        randomVisit = np.random.permutation(n) # generates the random order
        for elem in randomVisit:
            currentX = train[elem-1,:]
            prod = np.dot(currentX,w)
            y_j =trainlabel[elem][0]

            # if not match, update the weight vector
            if prod * (y_j) < 0:
                w = w + 0.1 * y_j* currentX.T
                
        # The checking code for after each full iteration of all the elements      
        allCorrect = True
        for check in range(n):
            curr = train[check,:]
            prod = np.dot(curr,w)
            if prod * (trainlabel[check][0]) < 0:
                allCorrect = False
                
        # If all data is correctly classified, we can stop looping
        if allCorrect:
#             print("break loop")
            break
                
                      
    return(w)
                
```
